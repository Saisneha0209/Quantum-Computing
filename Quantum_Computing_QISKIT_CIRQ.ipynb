```python
def factor(target):
    print(f"Factors for {target}:")
    n = 2
    while n * n <= target:
        if target % n == 0:
            m = target // n
            print(f" -> {n} * {m}")
        n += 1
    print("-" * 20)
factor(10)
```

```python
def check(n):
    if n < 2:
      return False
    if n == 2:
      return True
    if n % 2 == 0:
      return False
    for i in range(3, 1000, 2):
        if n % i == 0:
            return False
    return True

num = 1000000000000000003
if check(num):
    print(f"{num} is prime!")
else:
    print(f"{num} is definitely NOT a prime.")
```

```python
n = 1000000000000000003
x= int(n**0.5)
y = [i for i in range(3, x+ 1, 2) if n % i == 0]
if n % 2 == 0: print(f"{n} is Composite, factor: 2")
elif not y: print(f"{n} is Prime (checked up to {x})")
else: print(f"{n} is Composite, factors: {y}")
```
