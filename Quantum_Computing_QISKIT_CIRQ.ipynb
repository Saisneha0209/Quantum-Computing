```python
import qiskit #First Installed The Qiskit Library
from qiskit_aer import AerSimulator # Import AerSimulator

circuit=qiskit.QuantumCircuit(2)  #Now create a quantum circuit with two quibits
#lets apply hadamard gate to the first qubit
circuit.h(0)
creg=qiskit.ClassicalRegister(2 , name = 'c')
circuit.add_register(creg)
circuit.measure([0,1],[creg[0],creg[1]])
#SIMULATE THE CIRCUIT:-
simulator = AerSimulator() # Initialize the AerSimulator
result = simulator.run(circuit, shots=1024).result() # Use simulator.run instead of qiskit.execute
counts = result.get_counts()
print(counts)
```
